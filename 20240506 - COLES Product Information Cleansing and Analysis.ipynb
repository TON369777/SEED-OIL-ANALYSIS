## Further cleansing of the combined product info

```python
all_prod = pd.read_csv('AllCOLESPRODUCTINFO.csv')
```

```python
all_prod['Ingredients'] = all_prod['Ingredients'].fillna('')
```

```python
all_prod['Ingredients_lower_case'] = all_prod['Ingredients'].str.lower()
```

```python
all_prod['Ingredients_lower_case'] = all_prod['Ingredients_lower_case'].str.replace('oils', 'oil')
```

```python
# identify all specific seed oil names
seeds = ['vegetable oil', 'canola oil', 'soybean oil', 'sunflower oil', 'cottonseed oil', 
         'rapeseed oil', 'corn oil', 'safflower oil', 'peanut oil', 'sesame oil', 'grapeseed oil', 'flaxseed oil'
         ,'hempseed oil', 'pumpkin seed oil', 'almond oil', 'walnut oil'
        ]
```

```python
# introducing new columns to parse ingredients for seed oils
for seed in seeds:
    all_prod[seed] = all_prod['Ingredients_lower_case'].str.contains(seed).astype(int)
```

```python
all_prod['SEED_OIL_PRESENCE'] = all_prod[seeds].any(axis=1).astype(int)
```

```python
all_prod = all_prod.drop_duplicates(subset=['Product Name'])
```

```python
all_prod_PB = all_prod.drop(['Ingredients', 'Ingredients_lower_case'], axis=1)
```
